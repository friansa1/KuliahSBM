```python
import tensorflow as tf
import time

# Custom Attention Layer
class AttentionLayer1(tf.keras.layers.Layer):
    def __init__(self):
        super(AttentionLayer1, self).__init__()

    def build(self, input_shape):
        self.W_q = self.add_weight(name="W_q", shape=(input_shape[-1], input_shape[-1]), initializer="uniform", trainable=True)
        self.W_k = self.add_weight(name="W_k", shape=(input_shape[-1], input_shape[-1]), initializer="uniform", trainable=True)
        self.W_v = self.add_weight(name="W_v", shape=(input_shape[-1], input_shape[-1]), initializer="uniform", trainable=True)

    def call(self, inputs):
        q = tf.matmul(inputs, self.W_q)
        k = tf.matmul(inputs, self.W_k)
        v = tf.matmul(inputs, self.W_v)

        # Calculate attention scores
        attention_scores = tf.matmul(q, k, transpose_b=True)
        attention_scores = tf.nn.softmax(attention_scores, axis=-1)

        # Apply attention to values
        attention_output = tf.matmul(attention_scores, v)

        return attention_output
```

```python
# Custom function for quantization noise
def quantization_noise(x, num_bits=8):
    # Scale the values to the range [0, 2^num_bits - 1]
    scale = 2**num_bits - 1
    x_quantized = tf.round(x * scale) / scale
    return x_quantized
```

```python
def quantization_noise(x, quantization_factor=0.01):
    # Implement your quantization noise logic here using the quantization_factor
    # This is just a placeholder; adjust the logic based on your specific requirements.
    noise = tf.random.uniform(shape=tf.shape(x), minval=-quantization_factor, maxval=quantization_factor)
    return x + noise
```

```python
# Define a custom QuantizationNoise layer
class QuantizationNoiseLayer(Layer):
    def __init__(self, quantization_factor=0.01, **kwargs):
        super(QuantizationNoiseLayer, self).__init__(**kwargs)
        self.quantization_factor = quantization_factor
        self.uses_learning_phase = True

    def call(self, inputs, training=None):
        if training:
            quantization_noise = tf.random.uniform(shape=tf.shape(inputs), minval=-self.quantization_factor, maxval=self.quantization_factor)
            return inputs + quantization_noise
        return inputs
```

```python
# Custom function for speckle noise
import tensorflow as tf
from tensorflow.keras.layers import Layer

class QuantizationNoiseLayer(Layer):
    def __init__(self, num_bits=8, **kwargs):
        super(QuantizationNoiseLayer, self).__init__(**kwargs)
        self.num_bits = num_bits
        self.uses_learning_phase = True

    def call(self, inputs, training=None):
        if training:
            return quantization_noise(inputs, num_bits=self.num_bits)
        return inputs
```

```python
# Custom function for speckle noise
import tensorflow as tf
from tensorflow.keras.layers import Layer

class SpeckleNoiseLayer(Layer):
    def __init__(self, noise_factor=0.01, **kwargs):
        super(SpeckleNoiseLayer, self).__init__(**kwargs)
        self.noise_factor = noise_factor
        self.uses_learning_phase = True

    def call(self, inputs, training=None):
        if training:
            noise = inputs + tf.random.normal(shape=tf.shape(inputs), mean=0.0, stddev=self.noise_factor)
            return inputs * noise
        return inputs
```

```python
import statistics
import numpy
from math import sqrt

def mape(y, y_pred_all): 
    y, y_pred_all = np.array(y), np.array(y_pred_all)
    return np.mean(np.abs((y - y_pred_all) / y)) * 100
print('MAPE :',mape(y, y_pred_all) )

RMSE=sqrt(mean_squared_error(y,y_pred_all))
print('RMSE :',RMSE)

Avg = statistics.mean(y)
print("Mean :", Avg)

corr_matrix = numpy.corrcoef(y,y_pred_all)
corr = corr_matrix[0,1]
R_sq = corr**2

print("R2 :", R_sq)

CVRMSE = ((RMSE/Avg)*100)
print("CV-RMSE :", CVRMSE)

MBE = np.mean(y-y_pred_all) #here we calculate MBE
print("MBE :", MBE)

NMBE = ((MBE/Avg)*100)
print("NMBE :", NMBE)

VP = sqrt(np.mean((y_pred_all)**2))
#print('VP :', VP)

VA = sqrt(np.mean((y)**2))
#print('VA :', VA)

IC = (RMSE/(VP+VA))
print('Inequality Coefficient :', IC)
```
